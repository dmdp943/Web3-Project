```python
def get_dates():
    df = pd.read_csv("../raw_data/df_coinID_index_as_columns.csv")
    ico_dates = pd.read_parquet("../OLHC_Completo.parquet.gzip")
    coins_list = df.columns[1:].to_list()
    dates_dict = {}
    for coin in coins_list:
        try:
            first_date = ico_dates[f'{coin}'].dropna().index[0]
            data = ico_dates[f'{coin}'].loc[first_date:]
            dates_dict[f'{coin}'] = [data['Open'][0:100].fillna(method='ffill'), data['Close'][0:100].fillna(method='ffill')]
        except:
            pass
    df_final = pd.DataFrame.from_dict(dates_dict)
    return df_final.transpose()
```
