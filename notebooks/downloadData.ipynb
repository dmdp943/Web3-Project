```python
import pandas as pd
import requests
from datetime import datetime, timedelta
import time
format_date = lambda x: datetime.fromtimestamp(int(x)/1000)
```

## Download OLHC

```python
req = requests.get(f"https://api.coingecko.com/api/v3/coins/01coin/ohlc?vs_currency=usd&days=max")
resp = req.json()
dados = pd.DataFrame(resp)
dados.columns = ['Date','Open', 'High', 'Low', 'Close']
dados['Date'] = dados['Date'].apply(format_date)
dados
```

```python
req = requests.get("https://api.coingecko.com/api/v3/coins/list")
df = pd.DataFrame(req.json())
df
```

```python
res = {}
i = 0
for coin in df['id']:
    try:
        req = requests.get(f"https://api.coingecko.com/api/v3/coins/{coin}/ohlc?vs_currency=usd&days=max")
        resp = req.json()
        dados = pd.DataFrame(resp)
        dados.columns = ['Date','Open', 'High', 'Low', 'Close']
        dados['Date'] = dados['Date'].apply(format_date)
        dados.set_index('Date', inplace=True)

        dados.columns = pd.MultiIndex.from_product([[coin], dados.columns])
        res[coin] = dados
        dados.to_parquet(f'../raw_data/{coin}.parquet.gzip', compression='gzip')
    except Exception as e:
        print(coin)
    
    if i == 48:
        print('sleeping')
        time.sleep(100)
        print('waking')
        i = 0
        
    else: 
        i = i + 1
        
    
        
```

```python
historical_price = pd.concat([res[i] for i in res], axis=1)
historical_price
```

```python
historical_price.to_parquet('../raw_data/OLHC_Completo.parquet.gzip', compression='gzip')
```

## Download Historical IDO Price

```python
def getHist(coin, ini_date, delta_check=4):
    
    for i in range(1, delta_check+1):
        
        date_str = ini_date.strftime("%d-%m-%Y")        
        url = f'https://api.coingecko.com/api/v3/coins/{coin}/history?date={date_str}'         
        ini_date = ini_date + timedelta(1)
        req = requests.get(url)
        
        if req.status_code != 200:
            continue
        
        data = req.json()
        if "market_data" in data.keys():
            return data
        
        
```
