```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

# DAO



```python
all_dfs = []
for page in range(0,11):
    print(page)
    url = f'https://api.daomaker.com/getFundedCompanies?page={page}'
    content = requests.get(url)
    df = pd.DataFrame(content.json())
    all_dfs.append(df)
    
df_final = pd.concat(all_dfs)

df_final
```

```python
df_final.to_csv("daomakerinfo.csv")
```
