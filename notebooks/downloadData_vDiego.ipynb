```python
import pandas as pd
import requests
import json
from datetime import datetime
import time
format_date = lambda x: datetime.fromtimestamp(int(x)/1000)
```

```python
req = requests.get("https://api.coingecko.com/api/v3/coins/list")
df = pd.DataFrame(req.json())
df
```

```python
res = {}
i = 0
for coin in df['id']:
    try:
        req = requests.get(f"https://api.coingecko.com/api/v3/coins/{coin}")
        resp = req.json()
       # dados = pd.DataFrame(resp)
       # dados.columns = ['Date','Open', 'High', 'Low', 'Close']
       # dados['Date'] = dados['Date'].apply(format_date)
       # dados.set_index('Date', inplace=True)
       # dados.columns = pd.MultiIndex.from_product([[coin], dados.columns])
        res[coin] = resp
    except Exception as e:
        print(coin)
    
    if i == 49:
        time.sleep(120)
        i = 0
        
    else: 
        i = i + 1 
        
```

```python
#saving data of first 6749 coins in json
with open('data.json', 'w') as fp:
    json.dump(res, fp)
```

```python
#saving data of first 6749 coins in json
pd.DataFrame(res).to_csv("CoinDataID_6749.csv")
```
